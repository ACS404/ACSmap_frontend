## Section 1: The Two Models — A Tale of Theory vs. Practice

### 🧪 The OSI Model — The Academic Guideline (7 Layers)

The **Open Systems Interconnection (OSI) model** was developed by ISO in 1984 as a *conceptual framework* — a guideline for understanding and standardizing network communication. It was never meant to be implemented exactly as written; it's a teaching tool and a reference architecture.

```
┌─────────────────────────────────────────────────────┐
│  Layer 7 │ APPLICATION  │ HTTP, HTTPS, DNS, FTP      │
│  Layer 6 │ PRESENTATION │ SSL/TLS, Encryption, JPEG  │
│  Layer 5 │ SESSION      │ Session management, tokens │
│  Layer 4 │ TRANSPORT    │ TCP, UDP, port numbers      │
│  Layer 3 │ NETWORK      │ IP addresses, routing       │
│  Layer 2 │ DATA LINK    │ MAC addresses, Ethernet     │
│  Layer 1 │ PHYSICAL     │ Bits, cables, Wi-Fi signals │
└─────────────────────────────────────────────────────┘
```

**Key Mnemonic:** *"All People Seem To Need Data Processing"* (Application → Physical)

---

### 🚀 The TCP/IP Model — The Deployment Standard (5 Layers)

The **TCP/IP model** (also called the Internet Model) is what the internet *actually runs on*. It was developed by DARPA and became the practical standard. It collapses some OSI layers together:

```
┌─────────────────────────────────────────────────────────────────┐
│  Layer 5 │ APPLICATION  │ HTTP/HTTPS, DNS, SSH, SMTP            │
│          │              │ (Combines OSI Layers 5, 6, 7)         │
├──────────┼──────────────┼───────────────────────────────────────┤
│  Layer 4 │ TRANSPORT    │ TCP (reliable), UDP (fast)            │
│          │              │ = OSI Layer 4                         │
├──────────┼──────────────┼───────────────────────────────────────┤
│  Layer 3 │ NETWORK      │ IP addressing, packet routing         │
│          │              │ = OSI Layer 3                         │
├──────────┼──────────────┼───────────────────────────────────────┤
│  Layer 2 │ DATA LINK    │ Ethernet frames, MAC addresses        │
│          │              │ = OSI Layer 2                         │
├──────────┼──────────────┼───────────────────────────────────────┤
│  Layer 1 │ PHYSICAL     │ Electrical signals, fiber, radio waves│
│          │              │ = OSI Layer 1                         │
└─────────────────────────────────────────────────────────────────┘
```

### Key Difference Summary

| Attribute | OSI Model | TCP/IP Model |
|-----------|-----------|-------------|
| Layers | 7 | 5 |
| Purpose | Conceptual guideline | Practical deployment standard |
| Session Layer | Explicit (Layer 5) | Merged into Application |
| Presentation Layer | Explicit (Layer 6) | Merged into Application |
| Status | Academic reference | The actual internet |
| Developed by | ISO (1984) | DARPA (1970s) |

> **Bottom line:** Use OSI to *think* about networking. Use TCP/IP to *build* networking.

## Section 3: Encapsulation & Decapsulation

**Encapsulation** is the process of wrapping data with headers as it travels *down* the layers (sending side).
**Decapsulation** is stripping those headers as it travels *up* the layers (receiving side).

Think of it like mailing a letter:
- You write the letter (Application Layer data)
- Put it in an envelope with a return address (Transport Layer — TCP segment)
- Put that in a postal bag with a routing label (Network Layer — IP packet)
- Load it into a delivery truck (Data Link — Ethernet frame)
- The truck drives on a road (Physical Layer)

```
  SENDING (Down the Stack)        RECEIVING (Up the Stack)
  ========================        ========================
  [App Data]                      Strip DATA LINK header
       ↓ add TCP header                    ↓
  [TCP | App Data]                Strip NETWORK header
       ↓ add IP header                     ↓
  [IP | TCP | App Data]           Strip TRANSPORT header
       ↓ add Ethernet header               ↓
  [ETH | IP | TCP | App Data]     Read APPLICATION data ✓
       ↓ convert to bits
  101010101... (physical)
```

## Section 4: MTU — Maximum Transmission Unit

### What is MTU?

**MTU (Maximum Transmission Unit)** is the largest size of a single data packet that can be transmitted over a network link without being fragmented.

- Standard Ethernet MTU: **1500 bytes**
- Wi-Fi MTU: **2304 bytes** (but usually 1500 to match Ethernet)
- Internet backbone: often uses **Jumbo Frames** up to 9000 bytes

### Why does MTU exist?

Network hardware (routers, switches, NICs) can only handle packets up to a certain size. If your data is larger than the MTU, it gets **fragmented** — chopped into multiple packets.

### MTU and the ACS Body Map

When the ACS Body Map JavaScript loads the cancer data (the `BM_CANCERS` object with 80+ cancer descriptions), that data is *much* larger than 1500 bytes. Let's calculate how many packets it takes.

````python
import math

# ─── MTU CALCULATOR ───────────────────────────────────────────
# Simulating ACS Body Map data transmission

# Standard Ethernet MTU
ETHERNET_MTU = 1500  # bytes

# Headers eat into usable payload space:
IP_HEADER_SIZE  = 20   # bytes (IPv4)
TCP_HEADER_SIZE = 20   # bytes (minimum, no options)
TLS_OVERHEAD    = 29   # bytes (TLS 1.3 record header + MAC)

# Maximum Segment Size (MSS) = MTU - IP - TCP headers
MSS = ETHERNET_MTU - IP_HEADER_SIZE - TCP_HEADER_SIZE
# Usable payload with TLS encryption (HTTPS)
HTTPS_PAYLOAD = MSS - TLS_OVERHEAD

print("=" * 60)
print("  MTU BREAKDOWN")
print("=" * 60)
print(f"  Ethernet MTU        : {ETHERNET_MTU} bytes")
print(f"  - IP Header         : {IP_HEADER_SIZE} bytes")
print(f"  - TCP Header        : {TCP_HEADER_SIZE} bytes")
print(f"  = Max Segment Size  : {MSS} bytes")
print(f"  - TLS 1.3 Overhead  : {TLS_OVERHEAD} bytes")
print(f"  = HTTPS Payload     : {HTTPS_PAYLOAD} bytes per packet")

print()
print("=" * 60)
print("  ACS BODY MAP — DATA SIZE ESTIMATES")
print("=" * 60)

# Realistic size estimates for ACS Body Map assets
assets = [
    ("BM_CANCERS JS object (80+ cancers, descriptions)", 45_000),
    ("BM_HOTSPOTS + BM_SYSTEMS data",                     3_200),
    ("SVG body diagram (female)",                         28_000),
    ("CSS stylesheets (inline + external)",               18_500),
    ("Fonts (Cormorant Garamond subset)",                 24_000),
    ("Typical HTTPS API JSON response",                    4_500),
]

print(f"  {'Asset':<45} {'Size':>8} {'Packets':>8}")
print(f"  {'─'*45} {'─'*8} {'─'*8}")

total_bytes = 0
for name, size in assets:
    packets = math.ceil(size / HTTPS_PAYLOAD)
    total_bytes += size
    print(f"  {name:<45} {size:>7,}B {packets:>7} pkt")

total_packets = math.ceil(total_bytes / HTTPS_PAYLOAD)
print(f"  {'─'*45} {'─'*8} {'─'*8}")
print(f"  {'TOTAL':<45} {total_bytes:>7,}B {total_packets:>7} pkt")

print()
print(f"  At 100 Mbps: ~{(total_bytes * 8) / 100_000_000 * 1000:.1f} ms theoretical transfer time")
print(f"  Real-world with TCP handshake + TLS + RTT: typically 200-800ms")
````

## Section 5: TCP — Why Your Cancer Data Arrives Correctly

When the ACS Body Map fetches data, it uses **TCP** (Transmission Control Protocol). Here's why that matters:

### The TCP 3-Way Handshake

Before any cancer data moves, TCP establishes a reliable connection:

```
  Browser (Client)              GitHub Pages (Server)
       │                                │
       │  ── SYN ──────────────────►   │  "I want to connect"
       │                                │
       │  ◄──── SYN-ACK ──────────── │  "OK, I'm ready"
       │                                │
       │  ── ACK ──────────────────►   │  "Great, let's go!"
       │                                │
       │  ── GET /cancer/lung.html ──►  │  NOW data flows
       │                                │
       │  ◄──── [Packet 1 of 30] ───── │
       │  ◄──── [Packet 2 of 30] ───── │
       │  ◄──── [Packet 3 of 30] ───── │  (MTU fragmented)
       │         ...                    │
       │  ◄──── [Packet 30 of 30] ───  │
       │                                │
       │  ── ACK (all received) ─────►  │
```

**TCP guarantees:**
- Every packet arrives (retransmits if lost)
- Packets reassembled in the correct order
- Errors detected via checksum
- Flow control prevents overwhelming the receiver

> **Why not UDP?** UDP is faster (no handshake, no retransmit) but packets can arrive out of order or not at all. Fine for video streaming, terrible for displaying accurate cancer information.

````python
import random

# ─── TCP vs UDP SIMULATION ────────────────────────────────────
# Demonstrates why ACS Body Map uses TCP, not UDP

def simulate_tcp_transfer(data_kb, packet_loss_rate=0.05):
    """Simulate TCP: slower but 100% reliable"""
    total_packets = math.ceil(data_kb * 1024 / HTTPS_PAYLOAD)
    received = []
    retransmits = 0
    
    for seq in range(total_packets):
        # TCP retransmits until received
        while True:
            if random.random() > packet_loss_rate:  # packet gets through
                received.append(seq)
                break
            retransmits += 1  # TCP retransmits lost packet
    
    received.sort()  # TCP reorders packets
    return {
        "sent": total_packets,
        "received": len(received),
        "retransmits": retransmits,
        "complete": len(received) == total_packets,
        "in_order": received == list(range(total_packets))
    }

def simulate_udp_transfer(data_kb, packet_loss_rate=0.05):
    """Simulate UDP: faster but drops packets"""
    total_packets = math.ceil(data_kb * 1024 / HTTPS_PAYLOAD)
    received = []
    
    for seq in range(total_packets):
        if random.random() > packet_loss_rate:  # UDP just... doesn't retry
            received.append(seq)
        # dropped packet? UDP shrugs and moves on
    
    # UDP doesn't reorder — simulate arrival order shuffled
    random.shuffle(received)
    
    return {
        "sent": total_packets,
        "received": len(received),
        "dropped": total_packets - len(received),
        "complete": len(received) == total_packets,
        "in_order": False  # UDP doesn't guarantee order
    }

# ACS BM_CANCERS object ≈ 45KB
DATA_KB = 45
LOSS_RATE = 0.03  # 3% packet loss (typical for WiFi)

random.seed(42)
tcp_result = simulate_tcp_transfer(DATA_KB, LOSS_RATE)
udp_result = simulate_udp_transfer(DATA_KB, LOSS_RATE)

print("=" * 60)
print(f"  TRANSFERRING ACS BM_CANCERS DATA ({DATA_KB}KB) — 3% Packet Loss")
print("=" * 60)
print()
print("  TCP RESULT:")
print(f"    Packets sent     : {tcp_result['sent']}")
print(f"    Packets received : {tcp_result['received']}")
print(f"    Retransmissions  : {tcp_result['retransmits']}")
print(f"    Data complete?   : {'✅ YES' if tcp_result['complete'] else '❌ NO'}")
print(f"    In correct order?: {'✅ YES' if tcp_result['in_order'] else '❌ NO'}")
print(f"    Result           : Body Map displays ALL 80+ cancer types correctly")
print()
print("  UDP RESULT:")
print(f"    Packets sent     : {udp_result['sent']}")
print(f"    Packets received : {udp_result['received']}")
print(f"    Packets DROPPED  : {udp_result['dropped']}")
print(f"    Data complete?   : {'✅ YES' if udp_result['complete'] else '❌ NO — DATA MISSING'}")
print(f"    In correct order?: {'✅ YES' if udp_result['in_order'] else '❌ NO — OUT OF ORDER'}")
print(f"    Result           : Body Map renders BROKEN — some cancers missing or garbled")
print()
print("  → CONCLUSION: HTTPS (TCP) is the right choice for the ACS Body Map")
````

## Section 6: DNS — How the Browser Finds GitHub Pages

Before TCP even starts, your browser needs to know *where* `pages.opencodingsociety.com` is.

**DNS (Domain Name System)** is the internet's phonebook. It translates human-readable hostnames into IP addresses.

```
Step 1: Browser checks local cache
        → "Do I already know the IP for pages.opencodingsociety.com?"

Step 2: OS checks /etc/hosts
        → "Any local overrides?"

Step 3: Recursive DNS resolver (your ISP or 8.8.8.8)
        → "Let me ask the root servers..."

Step 4: Root Nameserver → ".com TLD server" → "opencodingsociety.com nameserver"
        → Returns: 185.199.108.153 (GitHub Pages IP)

Step 5: Browser now has IP → TCP connection begins
```

**In the ACS project context:**  
GitHub Pages uses a CNAME record so `pages.opencodingsociety.com` resolves to GitHub's CDN infrastructure. This means your cancer data is served from a server geographically close to each visitor — minimizing RTT (Round Trip Time) and speeding up the Body Map.

````python
import socket
import time

# ─── DNS RESOLUTION IN PYTHON ─────────────────────────────────
# Let's actually resolve some hostnames and measure lookup time

hosts_to_resolve = [
    ("pages.opencodingsociety.com", "ACS Body Map (GitHub Pages)"),
    ("www.cancer.org",               "American Cancer Society"),
    ("github.com",                   "GitHub (our deployment host)"),
    ("api.anthropic.com",            "Claude API (future AI feature)"),
]

print("=" * 65)
print("  DNS RESOLUTION — LAYER 5 (Application) OPERATION")
print("=" * 65)
print(f"  {'Host':<38} {'IP Address':<17} {'Time':>8}")
print(f"  {'─'*38} {'─'*17} {'─'*8}")

for hostname, description in hosts_to_resolve:
    try:
        start = time.time()
        ip = socket.gethostbyname(hostname)
        elapsed = (time.time() - start) * 1000
        print(f"  {hostname:<38} {ip:<17} {elapsed:>6.1f}ms")
        print(f"    └─ {description}")
    except Exception as e:
        print(f"  {hostname:<38} {'[FAILED]':<17}")
        print(f"    └─ {description}: {e}")

print()
print("  Note: DNS lookup time is separate from page load time.")
print("  Browsers cache DNS results to avoid this overhead on repeat visits.")
````

## Section 7: HTTPS & TLS — Keeping Cancer Data Private

The ACS Body Map uses **HTTPS** (HTTP Secure). This maps to:
- **OSI Layer 6 (Presentation)**: TLS encryption/decryption
- **TCP/IP Layer 5 (Application)**: Combined with HTTP

### Why HTTPS matters for health data:

Without HTTPS, an attacker could:
1. Read which cancer types a user is researching (privacy violation)
2. Inject fake cancer information into the page (integrity attack)
3. Steal any session tokens or form data

### TLS Handshake (simplified):

```
Browser                          GitHub Pages Server
   │                                    │
   │── ClientHello (supported ciphers) ►│
   │                                    │
   │◄─ ServerHello + Certificate ───── │
   │                                    │
   │── Verify cert (is it from a       │
   │   trusted CA like DigiCert?) ─────►│
   │                                    │
   │── Key exchange (ECDHE) ───────────►│
   │◄─ Finished ─────────────────────  │
   │                                    │
   │   [All data now AES-256 encrypted] │
   │── GET /cancer/lung.html ──────────►│  ← encrypted
```

The **TLS overhead** we subtracted in the MTU calculation comes from the encryption record headers and message authentication codes that wrap every encrypted payload.

## Section 9: How GitHub Pages Deploys the ACS Body Map

When you push code to the `main` branch of your GitHub Pages repository, here's the full network chain that makes it live:

```
1. git push origin main
   → Your computer connects to github.com:22 (SSH) or :443 (HTTPS)
   → TCP/IP carries your code diff in MTU-sized packets

2. GitHub Actions triggers
   → Jekyll builds the site (processes .html, .md, Liquid templates)
   → Outputs to GitHub's CDN edge nodes worldwide

3. GitHub Pages CDN distributes to edge nodes
   → San Diego user → Los Angeles edge node (low RTT ~3ms)
   → User in London → London edge node (would be ~130ms if served from US)

4. Visitor loads pages.opencodingsociety.com/cancer/body-map
   → DNS resolves to nearest GitHub Pages IP
   → TLS handshake establishes encrypted tunnel
   → HTTP/2 multiplexes HTML + CSS + JS + SVG assets simultaneously
   → Each asset fragmented into MTU-sized packets
   → Browser reassembles and renders the ACS Body Map
```

### Key Networking Concepts Used in This Deployment:

| Concept | Where it applies |
|---------|------------------|
| DNS | Resolving `pages.opencodingsociety.com` to GitHub IP |
| TCP/IP Layer 5 | HTTPS requests for HTML, CSS, SVG files |
| TCP/IP Layer 4 | TCP ensuring all cancer data packets arrive intact |
| TCP/IP Layer 3 | IP routing across the internet to GitHub's servers |
| MTU Fragmentation | 45KB BM_CANCERS object split into ~32 packets |
| TLS Encryption | Protecting user's health research privacy |
| CDN Edge Caching | GitHub Pages delivering assets from nearest server |

## Section 10: Summary

### OSI vs TCP/IP — The Core Distinction

| | OSI | TCP/IP |
|-|-----|--------|
| Layers | **7** | **5** |
| Type | **Guideline** | **Standard** |
| Session layer | Explicit | Merged into App |
| Presentation layer | Explicit | Merged into App |
| Real-world use | Teaching / troubleshooting | Actual internet |

### MTU Key Facts
- **Ethernet MTU = 1500 bytes**
- **MSS = MTU − IP header (20) − TCP header (20) = 1460 bytes**
- Data larger than MTU is **fragmented** into multiple packets
- TCP reassembles fragments in order; UDP may not
- The ACS Body Map's 45KB cancer data = ~32 TCP segments
